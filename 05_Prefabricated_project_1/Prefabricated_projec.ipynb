##### Обратите внимание на аббревиатуру 'tbd' в столбце с оценкой пользователей. Отдельно разберите это значение и опишите, как его обработать;

В столбце user_score присутствуют значения nan и tbd. tbd является абревиатурой для ещё не вышедших игр, проверим предроложение и построим график:

#### Возьмите данные за соответствующий актуальный период. Актуальный период определите самостоятельно в результате исследования предыдущих вопросов. Основной фактор — эти данные помогут построить прогноз на 2017 год.

В качестве актуального периода для построения прогноза будем рассматривать период с 2012 по 2016 годы - этот период соответствует периоду появления и начала продаж молодых платформ (3DS, PS4, PSV, WiiU, XOne). Также на него приходится чуть менее половины всего объема продаж возрастных платформ (PS3, Wii, X360) и менее четверти продаж игр для персональных компьютеров.

```python
# Обрезаем текущий данные до актуальных
data_acrual = data.loc[data['year_of_release'] >= 2012].copy()
data_acrual.head()
```

```python
# Возьмем данный отрезок для построения графика жизненного цикла
data_new = data.loc[data['year_of_release'] > 2004].copy()
data_new.head()
```

Далее определяем первый и последний годы жизни платформы, делаем расчет времени жизни

```python
# Группируем данные по платформам и годам выпуска, вычисляя суммарные продажи
df_platform_sales = data_new.groupby(['platform', 'year_of_release'])['general_game_sales'].sum().reset_index()

# Группируем суммарные продажи по платформам, убирая нулевые продажи
df_platform_total = df_platform_sales.groupby('platform')['general_game_sales'].sum().reset_index().query('general_game_sales > 0')

# Создаем столбец 'percent_1' с 1% от суммарных продаж, но не более 0.5
df_platform_total['percent_1'] = np.minimum(0.01 * df_platform_total['general_game_sales'], 0.5)

# Объединяем данные с оригинальными суммарными продажами
df_platform_total = df_platform_total.merge(df_platform_sales, on='platform')

# Фильтруем данные, оставляя только годы выпуска, начиная с 'percent_1'
df_platform_total = df_platform_total.query('year_of_release >= @df_platform_total.percent_1')

# Группируем данные по платформам и находим минимальный и максимальный год выпуска
df_platform_pivot = df_platform_total.groupby('platform')['year_of_release'].agg(['min', 'max']).reset_index()

# Переименовываем столбцы для ясности
df_platform_pivot.columns = ['platform', 'year_min', 'year_max']

# Вычисляем продолжительность жизни платформы
df_platform_pivot['lifespan'] = df_platform_pivot['year_max'] - df_platform_pivot['year_min']

# Выводим результат, сортируя по году завершения
df_platform_pivot.sort_values(by='year_max', ascending=False)
```

Возьмем только самые молодые платформы.

```python
# Отфильтровываем платформы, которые закончили свой жизненный цикл в 2016 году и имеют продолжительность не более 5 лет
actual_platforms_data = df_platform_pivot.loc[(df_platform_pivot['year_max'] == 2016) & (df_platform_pivot['lifespan'] <= 5)].sort_values(by='year_max', ascending=False)

# Выводим отфильтрованные данные
actual_platforms_data
```

Сначала построим график по тем платформам чей жизненный цикл не кончился вплоть до 2016.

Сформируем сводную таблиц продаж по платфрормам для каждого из представленных регионов: Европа (EU), Северная Америка (NA) и Япония (JP).

```python
# Используем метод .pivot_table() для создания сводной таблицы для продаж по регионам и платформам
df_platform_region = data_acrual.pivot_table(
    index='platform', 
    values=['na_sales', 'eu_sales', 'jp_sales'], 
    aggfunc='sum'
).reset_index()

# Выводим полученную сводную таблицу
df_platform_region
```

```python
def data_platform_region_sales(df_sales, agg_column, region_name, add_total=True):
    # Выбираем столбцы для агрегации и региона, сортируем данные по региону в убывающем порядке
    df = df_sales[[agg_column, region_name]].sort_values(by=region_name, ascending=False)
    
    # Переименовываем столбцы для более ясного понимания
    df.columns = [agg_column, 'sales']
    
    # Если флаг add_total равен True, добавляем категорию "Другие" и считаем ее общие продажи и процент
    if add_total:
        # Вычисляем процент продаж для каждой категории
        df = df.assign(percent=(df['sales'] / df['sales'].sum()) * 100)
        
        # Суммируем продажи для всех категорий, начиная с шестой
        other_sum = df.iloc[5:]['sales'].sum()
        
        # Суммируем процент продаж для всех категорий, начиная с шестой, и округляем до двух знаков после запятой
        other_percent = df.iloc[5:]['percent'].sum().round(2)
        
        # Создаем копию DataFrame, содержащую только топ-5 категорий
        df = df.iloc[:5].copy()
        
        # Добавляем строку "Другие" с общими продажами и процентом
        df.loc[5] = ['others', other_sum, other_percent]

    return df
```

```python
# Создаем словарь data_platform, в котором каждому региону (Северная Америка, Европа, Япония) соответствует DataFrame с продажами платформ
data_platform = {
    'Северная Америка' : data_platform_region_sales(df_platform_region, 'platform', 'na_sales'),
    'Европа' : data_platform_region_sales(df_platform_region, 'platform', 'eu_sales'),
    'Япония' : data_platform_region_sales(df_platform_region, 'platform', 'jp_sales'),
}
```

#### Влияет ли рейтинг ESRB на продажи в отдельном регионе?

Рассмотрим, как отличаются доли продаж игр различных возрастных категорий в представленных регионах.

Сформируем сводную таблиц продаж по возрастным категориям.

```python
# Создаем сводную таблицу для рейтингов игр по регионам с суммой продаж больше 2
df_esrb_region = data_acrual.pivot_table(
    index='rating', 
    values = ['na_sales', 'eu_sales', 'jp_sales'], 
    aggfunc='sum'
).reset_index().query('(na_sales + eu_sales + jp_sales) > 2')

# Преобразуем столбец 'rating' к типу 'object' для улучшения отображения данных
df_esrb_region.loc[:, 'rating'] = df_esrb_region['rating'].astype('object')

# Выводим полученные данные
df_esrb_region
```

```python
# Создаем функцию, которая формирует сводную таблицу для рейтингов игр по регионам с возможностью добавления категории 'others'
def data_rating_region_sales(df_sales, agg_column, region_name, add_total=True):
    # Сортируем данные по продажам в указанном регионе по убыванию
    df = df_sales[[agg_column, region_name]].sort_values(by=region_name, ascending=False)
    df.columns = [agg_column, 'sales']  # Переименовываем столбцы для ясного отображения данных

    if add_total:
        # Добавляем столбец 'percent' с процентом продаж от общей суммы
        df = df.assign(percent=(df['sales'] / df['sales'].sum()) * 100)
        # Вычисляем сумму и процент 'others' категории для элементов, которые не вошли в топ 5
        other_sum = df.iloc[5:]['sales'].sum()
        other_percent = df.iloc[5:]['percent'].sum().round(2)
        df = df.iloc[:5].copy()  # Копируем только топ 5 элементов
        df.loc[5] = ['others', other_sum, other_percent]  # Добавляем 'others' категорию

    return df
```

```python
# Создаем словарь, в котором ключами являются регионы, а значениями - данные по продажам игр в рейтинге ESRB
data_esrb_rates = {
    'Северная Америка' : data_rating_region_sales(df_esrb_region, 'rating', 'na_sales', False),
    'Европа' : data_rating_region_sales(df_esrb_region, 'rating', 'eu_sales', False),
    'Япония' : data_rating_region_sales(df_esrb_region, 'rating', 'jp_sales', False),
}
```

### Проверьте гипотезы

#### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;

Сформулируем нулевую гипотезу H₀:
"Средний пользовательский рейтинг игр для платформы Xbox One равен среднему пользовательскому рейтингу игр для платформы PC"

Альтернативная гипотеза H₁:
"Средний пользовательский рейтинг игр для платформы Xbox One отличается от среднего пользовательского рейтинга игр для платформы PC"

Критический уровень статистической значимости установим в 5%.

Проведем проверку гипотезы H₀.

```python
# Создаем списки пользовательских рейтингов для платформ Xbox One и PC
xone = list(data_acrual[(data_acrual['platform'] == 'xone')&(data_acrual['user_score'] != 0)]['user_score'])
pc = list(data_acrual[(data_acrual['platform'] == 'pc')&(data_acrual['user_score'] != 0)]['user_score'])

# Устанавливаем уровень статистической значимости (alpha)
alpha = 0.05

# Выполняем t-тест для сравнения пользовательских рейтингов на платформах Xbox One и PC
results = st.ttest_ind(
    xone, 
    pc,
    equal_var = True)

# Выводим p-значение, которое позволяет сделать вывод о статистической значимости различий
print('p-значение: {:.5f}'.format(results.pvalue))

# Проверяем результат теста и выводим соответствующее сообщение
if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу, рейтинги платформ Xbox One и PC разные")
else:
    print("Не получилось отвергнуть нулевую гипотезу, рейтинги платформ Xbox One и PC одинаковые")
```
