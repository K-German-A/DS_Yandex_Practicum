**Задание 4. Выведите количество пропущенных значений для каждого столбца. Используйте комбинацию двух методов.**

```python
# Вызываем метод 'isna()' для переменной 'data', который возвращает булеву маску,
# где True означает пропущенное (NaN) значение, а False - не пропущенное значение.
missing_values = data.isna()

# Затем вызываем метод 'sum()' для булевой маски 'missing_values', который суммирует количество True (пропущенных значений)
# для каждого столбца, так как True в булевой маске эквивалентно 1, а False - 0.
missing_values_count = missing_values.sum()

# После выполнения этой строки 'missing_values_count' будет содержать количество пропущенных значений для каждого столбца.
missing_values_count
```

**Задание 5. В двух столбцах есть пропущенные значения. Один из них — `days_employed`. Пропуски в этом столбце вы обработаете на следующем этапе. Другой столбец с пропущенными значениями — `total_income` — хранит данные о доходах. На сумму дохода сильнее всего влияет тип занятости, поэтому заполнить пропуски в этом столбце нужно медианным значением по каждому типу из столбца `income_type`. Например, у человека с типом занятости `сотрудник` пропуск в столбце `total_income` должен быть заполнен медианным доходом среди всех записей с тем же типом.**

```python
# Цикл по уникальным значениям столбца 'income_type' в DataFrame 'data'
for t in data['income_type'].unique():
    # Создаем булеву маску, которая проверяет, имеются ли пропущенные значения в столбце 'total_income'
    # для записей, у которых значение 'income_type' равно текущему уникальному значению 't'.
    missing_income_mask = (data['income_type'] == t) & (data['total_income'].isna())
    
    # Вычисляем медиану дохода ('total_income') для записей с текущим значением 'income_type' равным 't'.
    median_income = data.loc[data['income_type'] == t, 'total_income'].median()
    
    # Заполняем пропущенные значения в столбце 'total_income' средним значением (медианой) дохода для
    # записей с текущим значением 'income_type' равным 't'.
    data.loc[missing_income_mask, 'total_income'] = median_income

# В результате выполнения этого кода, пропущенные значения в столбце 'total_income' будут заменены на
# медианные значения дохода для соответствующей категории 'income_type'.
```

### Категоризация данных

**Задание 16. На основании диапазонов, указанных ниже, создайте в датафрейме `data` столбец `total_income_category` с категориями:**

- 0–30000 — `'E'`;
- 30001–50000 — `'D'`;
- 50001–200000 — `'C'`;
- 200001–1000000 — `'B'`;
- 1000001 и выше — `'A'`.


**Например, кредитополучателю с доходом 25000 нужно назначить категорию `'E'`, а клиенту, получающему 235000, — `'B'`. Используйте собственную функцию с именем `categorize_income()` и метод `apply()`.**

```python
# Определяем функцию 'categorize_income', которая принимает значение 'income' в качестве аргумента.
def categorize_income(income):
    try:
        # В блоке 'try' пытаемся выполнить следующие проверки и возврат категории дохода:
        if 0 <= income <= 30000:
            return 'E'  # Категория 'E' для дохода от 0 до 30,000
        elif 30001 <= income <= 50000:
            return 'D'  # Категория 'D' для дохода от 30,001 до 50,000
        elif 50001 <= income <= 200000:
            return 'C'  # Категория 'C' для дохода от 50,001 до 200,000
        elif 200001 <= income <= 1000000:
            return 'B'  # Категория 'B' для дохода от 200,001 до 1,000,000
        elif income >= 1000001:
            return 'A'  # Категория 'A' для дохода более 1,000,000
    except:
        pass  # Если возникнет ошибка, мы не делаем ничего

# Функция 'categorize_income' выполняет классификацию дохода в соответствии с заданными категориями.
# Важно убедиться, что 'income' передаваемое функции должно быть числом и не приводит к ошибкам.
```

```python
# Используем метод 'apply' для создания нового столбца 'total_income_category' в DataFrame 'data'.
# Мы применяем функцию 'categorize_income' к каждому элементу столбца 'total_income' и сохраняем результаты в новом столбце.
data['total_income_category'] = data['total_income'].apply(categorize_income)

# Этот код выполняет следующее действие:
# 1. Вызывает метод 'apply' для столбца 'total_income' в DataFrame 'data'.
# 2. В качестве аргумента для 'apply' передается функция 'categorize_income', которая будет применена к каждому элементу столбца 'total_income'.
# 3. Результатом выполнения 'apply' является новый столбец 'total_income_category', в котором каждый элемент будет содержать
#    категорию дохода, вычисленную с использованием функции 'categorize_income'.

# После выполнения этого кода, DataFrame 'data' будет содержать новый столбец 'total_income_category',
# в котором каждая запись будет содержать категорию дохода, соответствующую доходу из столбца 'total_income'.
```

**Задание 18. Создайте функцию, которая на основании данных из столбца `purpose` сформирует новый столбец `purpose_category`, в который войдут следующие категории:**

- `'операции с автомобилем'`,
- `'операции с недвижимостью'`,
- `'проведение свадьбы'`,
- `'получение образования'`.

**Например, если в столбце `purpose` находится подстрока `'на покупку автомобиля'`, то в столбце `purpose_category` должна появиться строка `'операции с автомобилем'`.**

**Используйте собственную функцию с именем `categorize_purpose()` и метод `apply()`. Изучите данные в столбце `purpose` и определите, какие подстроки помогут вам правильно определить категорию.**

```python
# Определяем функцию 'categorize_purpose', которая принимает строку 'row' в качестве аргумента.
def categorize_purpose(row):
    try:
        # В блоке 'try' пытаемся выполнить следующие проверки и возврат категории цели кредита:
        if 'автом' in row:
            return 'операции с автомобилем'  # Категория для цели, связанной с автомобилем.
        elif 'жил' in row or 'недвиж' in row:
            return 'операции с недвижимостью'  # Категория для цели, связанной с жильем или недвижимостью.
        elif 'свад' in row:
            return 'проведение свадьбы'  # Категория для цели, связанной с проведением свадьбы.
        elif 'образов' in row:
            return 'получение образования'  # Категория для цели, связанной с получением образования.
    except:
        return 'нет категории'  # Если возникнет ошибка, возвращаем "нет категории".

# Функция 'categorize_purpose' выполняет классификацию цели кредита в соответствии с заданными категориями.
# Важно убедиться, что 'row' является строкой, чтобы избежать ошибок.
```

```python
# Используем метод 'apply' для создания нового столбца 'purpose_category' в DataFrame 'data'.
# Мы применяем функцию 'categorize_purpose' к каждой строке столбца 'purpose' и сохраняем результаты в новом столбце.
data['purpose_category'] = data['purpose'].apply(categorize_purpose)

# Этот код выполняет следующее действие:
# 1. Вызывает метод 'apply' для столбца 'purpose' в DataFrame 'data'.
# 2. В качестве аргумента для 'apply' передается функция 'categorize_purpose', которая будет применена к каждой строке столбца 'purpose'.
# 3. Результатом выполнения 'apply' является новый столбец 'purpose_category', в котором каждая строка будет содержать
#    категорию цели кредита, вычисленную с использованием функции 'categorize_purpose'.

# После выполнения этого кода, DataFrame 'data' будет содержать новый столбец 'purpose_category',
# в котором каждая запись будет содержать категорию цели кредита, вычисленную с использованием функции 'categorize_purpose'.
```
